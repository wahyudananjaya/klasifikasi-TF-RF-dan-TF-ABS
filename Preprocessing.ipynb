```python
import pandas as pd
import numpy as np

df = pd.read_excel('dataa.xlsx')
df.head()
```

```python
df = pd.DataFrame(df[['Kelas','Teks']])
```

```python
df
```

```python
#melakukan case folding
df['Teks'] = df['Teks'].str.lower()

print('Hasil Case Folding : \n')
print(df['Teks'].head())
```

```python
import string
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize


# menghapus tanda baca
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))
df['Teks'] = df['Teks'].apply(remove_punctuation)

def remove_number(text):
    return re.sub(r"\d+", "", text)
df['Teks'] = df['Teks'].apply(remove_number)

def remove_character_special(text):
    # hapus titik, petik
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\'," ").replace('bit.ly'," ")
    # hapus mention, hastag dan link
    text = ' '.join(re.sub("([''""@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # hapus URL
    return text.replace("https://"," ").replace("http://", " ").replace("https*", " ").replace("bitly*", " ")


df['Teks'] = df['Teks'].apply(remove_character_special)

#melakukan tokenizing
def word_tokenize_wrapper(text):
    return word_tokenize(text)
df['teks_token'] = df['Teks'].apply(word_tokenize_wrapper)

print('Hasil Tokenizing')
print(df['teks_token'].head(20))
```

```python
from nltk.probability import FreqDist

def freqDist_wrapper(text):
    return FreqDist(text)

df['teks_token_frekuensi'] = df['teks_token'].apply(freqDist_wrapper)

print('HASIL FREKUENSI :\n')
print(df['teks_token_frekuensi'].head())
```

```python
df.head()
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")

list_stopwords = stopwords.words('indonesian')

list_stopwords.extend(["yg","hai...hai","ny","ya","loh","hai","nih"])

list_stopwords = set(list_stopwords)

# hapus stopword pada token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

df['teks_filtering'] = df['teks_token'].apply(stopwords_removal)

print('hasil filtering: \n')
print(df['teks_filtering'].to_string())
```

```python
df.head()
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

#membuat stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

#melakukan stemming
def stemmed_wrapper(term):
    return stemmer.stem(term)
term_dict = {}

for document in df['teks_filtering']:
    for term in document:
        if term not in term_dict: 
            term_dict[term] = ' '
            

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)

#memasukkan hasil ke dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['teks_stemming'] = df['teks_filtering'].swifter.apply(get_stemmed_term)

print('Hasil Stemming :\n')
print(df['teks_stemming'])
```
